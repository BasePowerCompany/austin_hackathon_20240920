```python
import matplotlib.pyplot as plt

# Plot outage instances to get a sense for the data.
df[['outage_duration_mins', 'customers_out']].plot.scatter(
    x='outage_duration_mins',
    y='customers_out',
    s=1
)

# Set bounds on the plot to make the data easier to view.
plt.ylim(bottom=0, top=10000)
plt.xlim(left=0, right=1440)
plt.show()
```
